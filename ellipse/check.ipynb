```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
f = 2800
WD = 30
a = 2800.060626
b = 18.42575822
r = 20
```

```python
theta = np.linspace(math.pi / 8.0, math.pi, 1000)
```

```python
path_length = np.hypot(a * np.cos(theta) + f, b * np.sin(theta)) + np.hypot(a * np.cos(theta) - f + WD, b * np.sin(theta) - r)
```

```python
path_length = np.hypot(a * np.cos(theta) + f, b * np.sin(theta) - 100) + np.hypot(a * np.cos(theta) - f + WD, b * np.sin(theta) - r - 100)
plt.plot(theta, path_length)
```

```python
coef_0 = f * f / (a * a) - 1.0;
coef_1 = - 2.0 * f * WD / (a * a);
coef_2 = WD * WD / (a * a) + r * r / (b * b);

print(coef_0, coef_1, coef_2)

t = (- coef_1 + math.sqrt(coef_1 ** 2 - 4.0 * coef_0 * coef_2)) / 2.0 / coef_2
t_minus = (- coef_1 - math.sqrt(coef_1 ** 2 - 4.0 * coef_0 * coef_2)) / 2.0 / coef_2
print(t, t_minus)
print(f - t * WD)
print(t * r)
supposed_theta = math.pi - math.acos((f - t * WD) / a)
print(supposed_theta)
```

```python
rs = np.linspace(0.001, r, 1000)
coef_0 = f * f / (a * a) - 1.0;
coef_1 = - 2.0 * f * WD / (a * a);
coef_2 = WD * WD / (a * a) + rs * rs / (b * b);

t = (- coef_1 + np.sqrt(coef_1 ** 2 - 4.0 * coef_0 * coef_2)) / 2.0 / coef_2
t_minus = (- coef_1 - np.sqrt(coef_1 ** 2 - 4.0 * coef_0 * coef_2)) / 2.0 / coef_2

supposed_theta = np.pi - np.arccos((f - t * WD) / a)

plt.plot(rs, t)
plt.figure()
plt.plot(rs, t_minus)
plt.figure()
plt.plot(rs, supposed_theta)
```

```python
normal_x = b * np.cos(theta) / math.hypot(a, b)
normal_y = a * np.sin(theta) / math.hypot(a, b)

cos_left = (normal_x * (-f - a * np.cos(theta)) + normal_y * (-np.sin(theta))) / np.hypot(-f - a * np.cos(theta), -np.sin(theta))
cos_right = normal_x * (f - WD - a * np.cos(theta)) + normal_y * (r - b * np.sin(theta)) / np.hypot(f - WD - a * np.cos(theta), r - b * np.sin(theta))
```

```python
plt.plot(theta, cos_left)
plt.plot(theta, cos_right)
```
