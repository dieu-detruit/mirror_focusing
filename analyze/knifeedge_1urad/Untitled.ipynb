```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y, profile = np.loadtxt('data/dev_0nrad/profile.txt').transpose()
plt.plot(y, profile)
```

```python
diff = np.diff(profile)
plt.plot(diff)
```

```python
cumsum = np.cumsum(profile)
plt.plot(cumsum)
```

```python
avr = np.average(profile)

y_rev = np.flip(y)
plt.plot(y, profile - avr)
plt.plot(y_rev, -profile + avr)
```

```python
norm_list = []

for folder in [0, 99, 199, 299, 399, 499, 599, 699,799, 900, 999]:
    y, profile = np.loadtxt('data/dev_{}nrad/profile.txt'.format(folder)).transpose()
    avr = np.average(profile)

    profile_rev = - np.flip(profile)
    
    profile_zeromean = profile - avr
    profile_rev_zeromean = profile_rev + avr
    difference = np.sum((profile_zeromean - profile_rev_zeromean)**2)
    norm_list.append(difference)
        
    plt.figure()
    plt.plot(y, profile_zeromean)
    plt.plot(y, profile_rev_zeromean)
    plt.plot(y, profile_zeromean - profile_rev_zeromean)
    difference = np.sum(profile_zeromean - profile_rev_zeromean)
```
